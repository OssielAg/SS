```python
def pC(a,b):
    a1,a2,a3 = a
    b1,b2,b3 = b
    return ((a2*b3-a3*b2),(a3*b1-a1*b3),(a1*b2-a2*b1))
def pP(a,b):
    a1,a2,a3 = a
    b1,b2,b3 = b
    return a1*b1+a2*b2+a3*b3
def to2D(v):
    x,y,z=v
    return (x,y)
```

```python
def cRecip(a,b,c):
    (a1,a2,a3) = a
    (b1,b2,b3) = b
    (c1,c2,c3) = c
    e = math.pi/(pP(a,pC(b,c)))
    (x,y,z) = pC(b,c)
    u = (x*e, y*e, z*e)
    (x,y,z) = pC(a,c)
    v = (x*e, y*e, z*e)
    (x,y,z) = pC(a,b)
    w = (x*e, y*e, z*e)
    return [u,v,w]
```
