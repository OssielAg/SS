```python
def damepardevecinos(red):
    ldv = dameVecinos(red)
    a = ldv[0]
    for v in ldv:
        ang = cAng(a,v)
        if round(ang,1)>0.0:
            if round(ang,1)<90.1:
                return a, v
    return (0,0), (0,0)
                
```
